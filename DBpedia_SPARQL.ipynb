```python
from SPARQLWrapper import SPARQLWrapper, JSON
```

```python
def get_tourist_attractions(latitude, longitude):
    # Формируем SPARQL-запрос
    sparql = SPARQLWrapper("http://dbpedia.org/sparql")
    sparql.setQuery("""
        SELECT DISTINCT ?place ?placeLabel ?description ?lat ?long ?category (COALESCE(?typeLabel, "Unknown") AS ?typeLabel)
        WHERE {
            ?place geo:lat ?lat ;
                  geo:long ?long ;
                  rdf:type ?type ;
                  rdfs:label ?placeLabel ;
                  dbo:abstract ?description ;
                  dcterms:subject ?category .

            OPTIONAL { ?place rdf:type ?type . }

            FILTER(lang(?placeLabel) = "en" && lang(?description) = "en")

            ?category skos:broader{0,14} <http://dbpedia.org/resource/Category:Tourist_attractions>.

            FILTER(?lat - """ + str(latitude) + """ <= 1.0 && """ + str(latitude) + """ - ?lat <= 1.0 &&
                  ?long - """ + str(longitude) + """ <= 1.0 && """ + str(longitude) + """ - ?long <= 1.0)
        }
        ORDER BY ?placeLabel
    """)

    sparql.setReturnFormat(JSON)
    results = sparql.query().convert()

    attractions = []
    for result in results["results"]["bindings"]:
        attraction = {
            "name": result["placeLabel"]["value"],
            "type": result["typeLabel"]["value"],
            "description": result["description"]["value"],
            "latitude": result["lat"]["value"],
            "longitude": result["long"]["value"],
            "iri": result["place"]["value"]
        }
        attractions.append(attraction)

    return attractions
```

```python
# Пример использования
latitude = 50.225141  # широта
longitude = 12.881876  # долгота
attractions = get_tourist_attractions(latitude, longitude)
attractions
```
